# Question 1.  Q1. Create a CSV file containing product names and prices. Read the data using Pandas. Use NumPy to calculate the average price. Write a function to identify products priced above the average and display them.

```python
import pandas as pd

data = {
    "Product": ["Pen", "Book", "Bag", "Bottle", "Scale"],
    "Price": [20, 120, 800, 250, 50]
}

df = pd.DataFrame(data)
df.to_csv("products.csv", index=False)
print("CSV file created successfully!")
```

```python
df = pd.read_csv("products.csv")
print(df)
```

```python
import numpy as np

avg = np.mean(df["Price"])

def above_average():
    print("Average Price =", avg)
    print("\nProducts above average price:")
    print(df[df["Price"] > avg])

above_average()
```

# Question 2. Q2. Read daily temperature data from a CSV file using Pandas. Use NumPy to calculate the highest, lowest, and average temperatures. Write functions to classify each day as:
 Cold (&lt;20°C)
 Moderate (20–30°C)
 Hot (&gt;30°C)
Display the category for each day.



```python
import pandas as pd

data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Temperature": [18, 25, 34, 29, 15, 32, 21]
}

df = pd.DataFrame(data)

df.to_csv("temperature.csv", index=False)

print(df)
print("\ntemperature.csv created successfully!")
```

```python
df = pd.read_csv("temperature.csv")
print(df)
```

```python
import numpy as np

print("Highest Temperature =", np.max(df["Temperature"]))
print("Lowest Temperature =", np.min(df["Temperature"]))
print("Average Temperature =", np.mean(df["Temperature"]))
```

```python
def category(temp):
    if temp < 20:
        return "Cold"
    elif temp <= 30:
        return "Moderate"
    else:
        return "Hot"

df["Category"] = df["Temperature"].apply(category)

print(df)
```
